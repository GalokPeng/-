```python
#导入相关库
import csv
import matplotlib
import matplotlib.pyplot as plt
import pandas as pd

#读取文件
nd = pd.read_csv("ncov.csv")
#print(nd)

#选取每个月月底的累计确诊人数，并分别存储到y1 – y5 中
for y1 in nd.loc[nd['日期'] == '2020/1/31',['累计确诊']].累计确诊:
    break
for y2 in nd.loc[nd['日期'] == '2020/2/29',['累计确诊']].累计确诊:
    break
for y3 in nd.loc[nd['日期'] == '2020/3/31',['累计确诊']].累计确诊:
    break
for y4 in nd.loc[nd['日期'] == '2020/4/30',['累计确诊']].累计确诊:
    break
for y5 in nd.loc[nd['日期'] == '2020/5/3',['累计确诊']].累计确诊:
    break

#选取每个月月底的累计治愈人数，并分别存储在 z1 – z5 中    
for z1 in nd.loc[nd['日期'] == '2020/1/31',['累计治愈']].累计治愈:
    break
for z2 in nd.loc[nd['日期'] == '2020/2/29',['累计治愈']].累计治愈:
    break
for z3 in nd.loc[nd['日期'] == '2020/3/31',['累计治愈']].累计治愈:
    break
for z4 in nd.loc[nd['日期'] == '2020/4/30',['累计治愈']].累计治愈:
    break
for z5 in nd.loc[nd['日期'] == '2020/5/3',['累计治愈']].累计治愈:
    break

#分别将y组和z组数据放入values1和values2表中
values1 = [y1,y2,y3,y4,y5]
values2 = [z1,z2,z3,z4,z5]

#折线图x轴坐标各点名称存入x表中
x = ['1.31','2.29','3.31','4.30','5.03']

#折线图名称
plt.title('The number of people diagnosed and cured',fontsize=15)
# x 标签
plt.xlabel('time',color='blue')
# y 标签
plt.ylabel('number of people',color='blue')
#画出累计确诊折线
plt.plot(x,values1,label="diagnosed",color='r',linewidth=2.0)
#画出累计治愈折线
plt.plot(x,values2,label=" cured",color='g',linewidth=3.0)
#显示折线图网格
plt.grid(alpha=0.3,linestyle=':')
#显示折线标签
plt.legend()
#显示
plt.show()
#分析结果简要
print("从折线图中显示：红色代表全国累计确诊人数变化，绿色代表累计治愈人数变化")
print("我们可以看到国内疫情爆发以来每个月底的情况，从1月31日至2.29日为爆发期，随后疫情得到了控制")
print("治愈人数也迅速的增加")








```

```python
#湖北新增确诊人数之和
Hubei_add = sum(nd['湖北新增确诊'])
#其他地区新增确诊人数
Others_add = sum(nd['全国新增确诊']) - Hubei_add
#标签存储至labels1表
labels1 = ['Hubei_add','Others_add']
#湖北与其他地区新增确诊人数存储至values3表
values3 = [Hubei_add,Others_add]
#设置饼图展现颜色
colors1 = ['b','y']
#分离饼图块
explode = [0.1,0]
#饼图标题
plt.title("The proportion of newly increased areas in China")
#画出饼图，并按百分比显示数据
plt.pie(values3,labels = labels1,explode = explode,colors = colors1,
       startangle = 180,
       shadow = True,autopct='%1.1f%%')
plt.axis('equal')
#显示饼图
plt.show()
#通过饼图得出需要的结论
print("全国湖北与其他地区累计新增确诊人数占比")
print("通过饼图，很明显看到，湖北是全国受病毒伤害最严重的地区")
```

```python
#取出每月月底现有重症人数，并存储至w1-w5中
for w1 in nd.loc[nd['日期'] == '2020/1/31',['现有重症']].现有重症:
    break
for w2 in nd.loc[nd['日期'] == '2020/2/29',['现有重症']].现有重症:
    break
for w3 in nd.loc[nd['日期'] == '2020/3/31',['现有重症']].现有重症:
    break
for w4 in nd.loc[nd['日期'] == '2020/4/30',['现有重症']].现有重症:
    break
for w5 in nd.loc[nd['日期'] == '2020/5/3',['现有重症']].现有重症:
    break
#x轴个点名称
x2 = ['1.31','2.29','3.31','4.30','5.03']
#柱形图标题
plt.title("The monthly number of severe patients in China")
#将w组存入values4表中
values4 = [w1,w2,w3,w4,w5]
#画出柱状图
plt.bar(x2,values4)
#显示
plt.show()
#通过柱状图得出结论
print("通过柱状图，我们很明显的看到，2月重症患者人数是最高的")
```
