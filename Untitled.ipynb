```python
import requests
import re
import os
import json

os.remove('新冠疫情各省数据.csv') 
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/73.0.3683.86 Safari/537.36'
}
r = requests.get('https://voice.baidu.com/act/newpneumonia/newpneumonia/?from=osari_aladin_banner', headers=headers)
r.encoding = r.apparent_encoding
match = '<script type="application/json" id="captain-config">(.*?)</script>'
data = re.findall(match, r.text, re.S)
js = json.loads(data[0])

with open(r'新冠疫情各省数据.csv', 'a', encoding='utf-8')as f:
    f.write('省份,累积确诊,累积治愈,现有确诊,新增确诊,累积死亡\n')
    f.close()

with open(r'新冠疫情各省数据.csv', 'a', encoding='utf-8')as f:
    for i in js['component'][0]['caseList']:
        shengfen = i['area']
        ljqz = i['confirmed']
        ljzy = i['crued']
        xyqz = i['curConfirm']
        xzqz = i['confirmedRelative']
        ljsw = i['died']
        # print(shengfen, ljqz, ljzy, xyqz, xzqz, ljsw)
        f.write(
            '{},{},{},{},{},{}\n'.format(shengfen,ljqz,ljzy,xyqz,xzqz,ljsw))

```

```python
import pandas as pd
data = pd.read_csv('新冠疫情各省数据.csv')
data.head(10)
```

```python
value1 = data['累积确诊']
value2 = data['现有确诊']
value3 = data['累积治愈']
value4 = data['累积死亡']
value1.head(10)
```

```python
v1=data['累积确诊'].sum() #计算全国累积确诊人数
v2=data['现有确诊'].sum() #计算全国现有确诊人数
v3=data['累积治愈'].sum() #计算全国已治愈人数
v4=data['累积死亡'].sum() #计算全国总死亡人数
# print(v1/34,v2/34,v3/34,v4/34)#各平均数
print(v2+v3+v4)
print(v1)#判断数据
```

```python
from pyecharts import Pie
from pyecharts import online
online()
attr = ['确诊','已治愈','已死亡']
vvz = [v2,v3,v4]
pie = Pie("疫情影响占比图")
pie.add("",attr,vvz)
pie
```
